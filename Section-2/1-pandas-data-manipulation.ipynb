```python
# lectura de datos
df = pd.read_csv('./datasets/Dataset Macro.csv', sep=';')  # The default separator is ","
# ver las 10 primeras líneas
df.tail(10) # para ver las 10 últimas líneas df.tail(10) 
```

```python
# Algunas otras cosas que podemos hacer:
#print(df.columns)    # nombres de las columnas
#print(df.info())     # información general del dataframe
#print(df.shape)      # (rows,columns)
#print(df.index)      # descripción de los índices
#df.count()    # contar el número de valores non-null
df.describe() # muestra estadísticas de las columnas 
```

```python
#Para analizar columnas particulares
print(df['country'].unique())       # List the categories of a particular column
print(len(df['country'].unique()))  # Count the categories of a particular column
print(df['country'].value_counts()) # Count the occurrences per category 
```

```python
# Se pueden crear nuevas columnas a partir de las iniciales
df['Nuevo Indice'] = df['GDP Growth'] * df['VIX Index'] / 100
df.head()
```

```python
# también se pueden borrar columnas
df = df.drop(['Nuevo Indice'], axis=1)
df.head()
```

```python
# podemos hacer nuevos dataframes que sean subconjuntos de columnas
df2 = df[['country','GDP Growth']]
df2.head()
```

```python
# podemos crear nuevos dataframes con filtros con valores de texto de las columnas 
df2 = df[df['country'] == 'Spain']
df2.head()
```

```python
# podemos crear nuevos dataframes con filtros con valores númericos de las columnas 
df2 = df[df['Unemployment Rate'] <= 5]
df2.head()
```

```python
# o combinar criterios
#criterio1 = df['country'] == 'Spain'
criterio1 = df['country'].isin(['Spain','France'])
criterio2 = df['Unemployment Rate'] <= 8
filtro = criterio1 & criterio2
df2 = df[filtro]
df2.head()
```

```python
# también es posible agrupar valores
# se puede hacer: mean(), sum(), count(), size(), max(), min(), var(), quantile([.25,.75])
#df.groupby('country').mean() # agrupando solo una columna
df.groupby(['year','country']).mean() # agrupando por varias columnas
```

```python
# gráficas básicas desde el dataframe: histogramas
#df.hist()
#df.hist(column='GDP Growth')
#df.hist(column='GDP Growth',by='country')
#df.hist(column='GDP Growth',by='country',figsize=[10, 6])
#df.hist(column=['GDP Growth','Public debt'])
df.hist(column=['GDP Growth','Public debt'],bins=25)
```

```python
# gráficas básicas desde el dataframe: gráfica x vs y (scatter plot)
#df.plot.scatter(x='year',y='GDP Growth')
#df.plot.scatter(x='year',y='GDP Growth',c='Unemployment Rate')
df.plot.scatter(x='year',y='GDP Growth',c='Unemployment Rate',colormap='viridis')
```
